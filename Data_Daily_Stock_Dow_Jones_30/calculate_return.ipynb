```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("^DJI.csv")
```

```python
test_df=df[df['Date']>'2015-12-30']
```

```python
test_df.head()
```

```python
df_price=test_df['Adj Close']
```

```python
df_date = test_df['Date']
```

```python
df_date.shape
```

```python
daily_return = df_price.pct_change(1)
```

```python
daily_return=daily_return[1:]
```

```python
daily_return.reset_index()
```

```python
daily_return.iloc[1]
```

```python
initial_amount = 10000
```

```python
total_amount=initial_amount
account_growth=list()
account_growth.append(initial_amount)
for i in range(len(daily_return)):
    total_amount = total_amount * daily_return.iloc[i] + total_amount
    account_growth.append(total_amount)
    
```
